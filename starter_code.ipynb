General code to read in the xlxs file you want from github

```python
import pandas as pd
import urllib.request
```

```python
# INSERT THE NAME OF THE FILE YOU WANT HERE THIS IS JUST AN EXAMPLE
file = 'aquatic_plants_lakes.xlsx'

url = f"https://github.com/haleyej/clean_water_corps_data/raw/main/{file}"
urllib.request.urlretrieve(url, file)
```

Our data is in excel files, so we use pd.read_excel. If for whatever reason you're trying to access data from the repo that's in a different format, just change this to the appropriate function (probably pd.read_csv).

```python
df = pd.read_excel(file)
df.head()
```
