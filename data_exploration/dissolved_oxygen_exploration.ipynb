```python
import pandas as pd
import urllib.request
import seaborn as sns
import numpy as np 
```

```python
file = 'dissolved_oxygen_temp_lake.xlsx'

url = f"https://github.com/haleyej/clean_water_corps_data/raw/main/{file}"
urllib.request.urlretrieve(url, file)
```

```python
df = pd.read_excel(file)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.Watershed.value_counts()
```

```python
df.describe(exclude=np.number)
```

```python
sns.scatterplot(x = 'Temp (°C)', y = 'DO Level (mg/L)', data = df)
```

<h2>Drop Outliers</h2>

```python
df = df[df['Temp (°C)'] <= 500]
```

```python
sns.scatterplot(x = 'Temp (°C)', y = 'DO Level (mg/L)', data = df)
```

```python
df = df[df['DO Level (mg/L)'] <= 200]
```

```python
sns.scatterplot(x = 'Temp (°C)', y = 'DO Level (mg/L)', data = df)
```

```python
df.groupby('County')['Lake Name'].count().sort_values(ascending = False)
```

```python
df.groupby('County')['Lake Name'].nunique().sort_values(ascending = False)
```
